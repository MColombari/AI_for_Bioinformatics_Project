```python
import matplotlib.pyplot as plt
import numpy as np
import json
import seaborn as sns
```

```python
PATH_TEST = "/work/h2020deciderficarra_shared/TCGA/OV/project_n16_data/GeneProcessedData/test_separation_regression.json"
PATH_TRAIN = "/work/h2020deciderficarra_shared/TCGA/OV/project_n16_data/GeneProcessedData/train_separation_regression.json"
```

```python
with open(PATH_TEST, 'r') as f:
    test_dict = json.load(f)
with open(PATH_TRAIN, 'r') as f:
    train_dict = json.load(f)

data_test = np.array(list(test_dict.values()))
data_train = np.array(list(train_dict.values()))

print(data_test)

# Plot the first histogram
sns.rugplot(data=data_test, height=0.5, label='test', color='blue')

# Plot the second histogram on the same axes
plt.hist(data_train, label='train', color='red')

# Add labels and title
plt.xlabel('OS')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.ylim(0, 80)
# Show the plot
plt.show()
```
