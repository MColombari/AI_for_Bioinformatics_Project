```python
import matplotlib.pyplot as plt
import numpy as np
import json
import seaborn as sns
```

```python
PATH_TEST = "/work/h2020deciderficarra_shared/TCGA/OV/project_n16_data/GeneProcessedData/test_separation_regression.json"
PATH_TRAIN = "/work/h2020deciderficarra_shared/TCGA/OV/project_n16_data/GeneProcessedData/train_separation_regression.json"
```

```python
with open(PATH_TEST, 'r') as f:
    test_dict = json.load(f)
with open(PATH_TRAIN, 'r') as f:
    train_dict = json.load(f)

data_test = np.array(list(test_dict.values()))
data_train = np.array(list(train_dict.values()))

tmp = list(test_dict.values())
tmp.sort()
print(tmp)

fig, ax1 = plt.subplots()

ax1.set_xlabel('OS')
ax1.set_ylabel('Frequency Train', color='red')
ax1.hist(data_train, bins= 30,label='train', color='red')
ax1.tick_params(axis='y', labelcolor='red')

ax2 = ax1.twinx()  # instantiate a second Axes that shares the same x-axis

ax2.set_ylabel('Frequency Test', color='blue')  # we already handled the x-label with ax1
sns.rugplot(data=data_test, height=0.5, label='test', color='blue', ax=ax2)
ax2.set_ylim([0, 2])
ax2.tick_params(axis='y', labelcolor='blue')

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
