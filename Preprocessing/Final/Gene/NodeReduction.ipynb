```python
import json
# Read data

with open("/work/h2020deciderficarra_shared/TCGA/OV/project_n16_data/GeneProcessedData/gene_variance_order_COMPLETE.json", 'r') as file:
    dict_gene = json.load(file)
```

```python
# importing the required libraries
import matplotlib.pyplot as plt
import numpy as np

set_focus = 2000

for f in dict_gene.keys():
    # define data values
    x = np.array([float(dict_gene[f][i][1]) for i in range(set_focus)])  # X-axis points
    y = np.array(range(1, set_focus + 1))

    plt.plot(x)  # Plot the chart
    plt.yscale("log")
    plt.title(f)
    plt.xlabel("Order")
    plt.ylabel("Variance value")
    plt.show()  # display
    # print([float(t[1]) for t in dict_gene[f]])
```

```python
# Do gene actually match between features?

index = 0
len_gene = None
while True:
    tmp_set = set()
    for f in dict_gene.keys():
        if len_gene == None:
            len_gene = len(dict_gene[f])
        tmp_set.add(dict_gene[f][index][0])
    if len(tmp_set) > 1:
        print(f"Found discrepancy at gene: {index + 1}")
        for f in dict_gene.keys():
            print(f"\t{f}: \"{dict_gene[f][index][0]}\"")
    index += 1
    if len_gene <= index:
        break
```
