```python
def parse_row(row):
    first = row.split(".")[1].split(" ")[0]
    second = row.split(".")[2].split(" ")[0]
    val = int(row.split(" ")[-1])

    return first, second, val
```

```python
# scored links between proteins
# https://string-db.org/cgi/download?sessionId=bClwVzYouNKE&species_text=Homo+sapiens&settings_expanded=0&min_download_score=0&filter_redundant_pairs=0&delimiter_type=txt

comparison_dict = {}
values_list = []

# Read file
with open('9606.protein.links.v12.0.txt', 'r') as file:
    skip_first = True
    for line in file:
        if skip_first:
            skip_first = False
            continue
        f, s, v = parse_row(line)
        if not f in comparison_dict.keys():
            comparison_dict[f] = {}
        comparison_dict[f][s] = v
        values_list.append(v)


```

```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np

print(f"Median of median {np.median(np.array(values_list))}")
print(f"Max of median {np.max(np.array(values_list))}")
print(f"Min of median {np.min(np.array(values_list))}")

plt.figure(figsize=(10, 6))
sns.histplot(np.array(values_list), bins=50, kde=True, color="red")
plt.title("Distribution of \'median for each node with eachother\'", fontsize=16)
plt.xlabel("value", fontsize=14)
plt.ylabel("Ammount per value", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
